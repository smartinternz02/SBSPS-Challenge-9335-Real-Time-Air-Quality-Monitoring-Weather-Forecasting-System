```python
hk_file = "../../data/hongkong_taiwan_aqi/hongkong_monitor_coords.csv"
hk = pd.read_csv(hk_file, skiprows=2)
```

```python
hk = hk[['ENGLISH NAME', 'LONGITUDE', 'LATITUDE']]
hk
```

```python
hk = hk.assign(id = [str(i) + 'HK' for i in range(16)])
```

```python
def convert(tudes):
    return np.array([sum(float(x) / 60 ** n for n, x in enumerate(tude.split('-'))) for tude in tudes])
```

```python
hk.loc[:,'LONGITUDE'] = convert(hk['LONGITUDE'].values)
```

```python
hk.loc[:,'LATITUDE'] = convert(hk['LATITUDE'].values)
```

```python
hk
```

```python
hk.to_csv("../../data/hongkong_taiwan_aqi/hongkong_monitor_coords_processed.csv")
```
